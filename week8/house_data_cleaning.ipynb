```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data from House_Data.csv
data = pd.read_csv('House_Data.csv')
df = pd.DataFrame(data)

# Display the first 5 rows of the data
print(df.head())
```

```python
# Handle missing values
print(df.isnull().sum())
```

```python
# Check the missing values in the society column
# print(df['society'].isnull().sum())
# Put 'Unknown' in place of missing values in the society column
df['society'] = df['society'].fillna('Unknown')
print(df['society'].isnull().sum())
```

```python
# Check the missing value for bath and balcony columns
# print(df['bath'].isnull().sum())
# print(df['balcony'].isnull().sum())

# total_sqft      
def convert_sqft_to_num(x):
    if '-' in x:
        nums = x.split('-')
        return (float(nums[0]) + float(nums[1])) / 2
    try:
        return float(x)
    except:
        return None

# total_sqft  
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)

# total_sqft    
df['sqft_range'] = pd.qcut(df['total_sqft'], q=3, labels=['Low', 'Medium', 'High'])

#    bath balcony   
mean_values = df.groupby('sqft_range').mean()

#      
df['bath'] = df.apply(lambda row: mean_values.loc[row['sqft_range'], 'bath'] if np.isnan(row['bath']) else row['bath'], axis=1)
df['balcony'] = df.apply(lambda row: mean_values.loc[row['sqft_range'], 'balcony'] if np.isnan(row['balcony']) else row['balcony'], axis=1)

```
