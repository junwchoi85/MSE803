## Scenario:
You are hired as a data analyst for a local environmental conservation organisation. Their mission is to
protect and restore the native fish populations in the Avon River. They have collected water quality data
(temperature, pH, dissolved oxygen) and fish population data (species, count, size) for several years at
different sites along the river.

```python
import pandas as pd

# Open CSV
fish_population = pd.read_csv('../data/fish_population.csv')
water_quality = pd.read_csv('../data/water_quality.csv')

# Print 10 first rows
print(fish_population.head(10))
print(water_quality.head(10))

# Handle missing values and convert data types if necessary
fish_population.fillna(0, inplace=True)
water_quality.fillna(0, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Visualize water quality data
water_quality.groupby('Site ID').mean(numeric_only=True).plot(kind='bar')
plt.title('Average Water Quality by Site')
plt.show()

# Visualize fish population data
fish_population.groupby('Site ID').sum(numeric_only=True)['Count'].plot(kind='bar')
plt.title('Total Fish Count by Site')
plt.show()
```

```python
# Correlation analysis
numeric_water_quality = water_quality.select_dtypes(include=['float64', 'int64'])
correlation = numeric_water_quality.corrwith(fish_population['Count'])
print(correlation)

# Analyze the impact of specific water quality conditions on fish population
# Example: Relationship between temperature and fish count
plt.scatter(water_quality['Temperature (°C)'], fish_population['Count'])
plt.title('Temperature vs Fish Count')
plt.xlabel('Temperature (°C)')
plt.ylabel('Fish Count')
plt.show()
```
